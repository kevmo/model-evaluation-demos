```python
from openai import OpenAI
openai_client = OpenAI()
```

```python
openai_model_name = "gpt-5.4-mini"

def llm(prompt):
    response = openai_client.responses.create(
        model=openai_model_name,
        input=prompt
    )
    return response.output_text

llm("Is this connection working?")
```
